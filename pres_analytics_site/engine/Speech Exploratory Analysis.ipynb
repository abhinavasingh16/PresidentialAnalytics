```python
import pandas as pd
import numpy as np
```

```python
with open('raw/speech/debate.txt','r') as debate_file1:
    debate1=debate_file1.read()
with open('raw/speech/debate2.txt','r') as debate_file2:
    debate2=debate_file2.read()
```

```python
debate1 = debate1.replace('(APPLAUSE)','')
debate2 = debate2.replace('(APPLAUSE)','')
```

```python
from stop_words import get_stop_words
stop_words = get_stop_words('english')
stop = stop_words
def annotate(debate_text):
    annotations = []
    i = -1
    for word in debate_text.split():
        if word.isupper() and word[-1] == ':':
            i+=1
            annotations.append([word,[]])
        else:
            annotations[i][1].append(word)
    for element_index in range(len(annotations)):
        annotations[element_index][1] = [w for w in annotations[element_index][1] if w not in stop]
        annotations[element_index][1] = ' '.join(annotations[element_index][1])
    return annotations

debates = annotate(debate1)+annotate(debate2)
debates
```

```python
df = pd.DataFrame(columns=['name','rebuttal','len_rebuttal'])
people = []
rebuttal = []
len_rebuttal = []
for speech in debates:
    person =speech[0].strip(':')
    people.append(person[0]+person[1:].lower())
    r = speech[1]
    rebuttal.append(speech[1])
    len_rebuttal.append(len(r.split()))
df['name'] = people
df['rebuttal'] = rebuttal
df['len_rebuttal'] = len_rebuttal
df.to_csv('debate_feature.csv')
```
